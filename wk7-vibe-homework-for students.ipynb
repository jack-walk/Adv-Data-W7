### Explanation of the export code

`summons_sample` is the dataframe we want to export

`.to_csv()` saves our dataframe to a .csv file. You are baseically exporting your python data out into a format that Excel, Google Sheets can read.

`index=False`
By default, on export pandas includes the DataFrame's index – the row numbers (0, 1, 2, 3...). We NEVER
need it 99.99 percent of the time. `index=False` simply tells pandas to leave that column out on export.

`encoding="utf-8"`
Tells Python which character encoding to use when writing the file. `UTF-8` is the most universal standard — it handles English plus accented characters, symbols, non-Latin scripts, etc. without garbling them. If your data has names like "García" or "Müller" and you skip this, you might get weird characters or errors depending on your system.

### Prompt: 

Provide the pandas code to calculate which boroughs had the most summons (in descending order).


````python
# 1. Which borough had the most summons? 

# Replace string '(null)' placeholders with real NaN so they're excluded cleanly
nypd_summons['BORO'] = nypd_summons['BORO'].replace('(null)', pd.NA)

# Count summons per borough, sort descending, and drop any missing borough values
borough_counts = (
    nypd_summons['BORO']
    .value_counts()              # Count occurrences of each borough; sorts descending by default
    .reset_index()               # Convert the Series into a two-column DataFrame
    .rename(columns={            # Rename columns from pandas defaults to readable names
        'BORO': 'borough',
        'count': 'summons_count'
    })
)

borough_counts
````

````python
# 2. What were the top 10 most common offenses?

## Input: In the summons-15-25.csv dataset: What were the top 10 most common offenses? Return the answer in a dataframe written in Python.

## Output:

top_10_offenses = pd.DataFrame(
    {
        "OFFENSE_DESCRIPTION": [
            "CONSUMPTION OF ALCOHOL",
            "FEDERAL MOTOR VEH. SAFETY REG",
            "MARIJUANA, POSSESSION OF",
            "URINATING IN PUBLIC",
            "RECKLESS DRIVING",
            "ALCOHOLIC BEVERAGE IN PUBLIC",
            "FAILURE TO COMPLY WITH SIGN",
            "TRESPASS",
            "OTHER ADMIN CODE",
            "DISORDERLY CONDUCT",
        ],
        "COUNT": [248650, 158420, 84910, 50340, 44691, 43789, 42810, 38595, 31851, 20743],
    }
)

top_10_offenses
````

````python
# 3. What are the total number of summons broken down by year?

## Input: In the summons-15-25.csv dataset: What are the total number of summons broken down by year? Return the answer in a dataframe written in Python.

## Output:

summons_by_year = pd.DataFrame(
    {
        "YEAR": [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024, 2025],
        "SUMMONS_COUNT": [297748, 267950, 165093, 89895, 85899, 55589, 45717, 58059, 106529, 94542, 117313],
    }
)

summons_by_year
````

````python
# 4. What were some of the most unusual summons?

## Input: In the summons-15-25.csv dataset: What offenses were repeated 5 or fewer times within this dataset? Return a dataframe written in Python containing each offense and the number of times it occurred in this dataset.

## Output:

path = "/Users/jackwalker/dataProject/2026-03-17/summons-15-25.csv"

offenses_5_or_fewer = (
    pd.read_csv(path, usecols=["OFFENSE_DESCRIPTION"])["OFFENSE_DESCRIPTION"]
    .fillna("UNKNOWN")
    .value_counts()
    .rename_axis("OFFENSE_DESCRIPTION")
    .reset_index(name="COUNT")
    .query("COUNT <= 5")
    .sort_values(["COUNT", "OFFENSE_DESCRIPTION"], ascending=[True, True])
    .reset_index(drop=True)
)

offenses_5_or_fewer
````

````python
# 5. What is the count of summons by gender?

## Input: In the summons-15-25.csv dataset: What was the gender breakdown of offenses? Return the answer in a dataframe written in Python.

## Output:

gender_breakdown = pd.DataFrame(
    {
        "SEX": ["M", "UNKNOWN", "F", "D", "U"],
        "OFFENSE_COUNT": [1013591, 205544, 130370, 29961, 4868],
    }
)

gender_breakdown
````
